```python
from sympy import *
```

In this notebook, $\kappa_s(\rho)$ and $E_0(\rho)$ as well as their derivatives are compared. First, we define the symbols and the symbollic functions:

```python
rho, n, x, y, s, alpha, E = symbols('rho n x y s alpha \mathcal{E}')
Q = Function("Q")(x)
W = Function("W")(y,x)
```

In the results, $W(y,x)$ corresponds to $W(y|x)$ in the derivations.

##$\kappa_s(\rho)$

```python
k = log(Integral(summation(Q * pow(W, 1-s*rho), (x, 0, n)) * pow(summation(Q * pow(W, s), (x, 0, n)), rho), (y, -oo, oo)))
```

```python
k
```

This result corresponds to $\kappa_s(\rho)$.

```python
k_rho = k.subs(s, 1/(1+rho))
```

```python
k_rho = simplify(k_rho)
k_rho = simplify(k_rho)
k_rho
```

This result corresponds to $\kappa_s(\rho)|_{s=\frac{1}{1+\rho}} = -E_0(\rho)$.

##$\frac{\partial \kappa_s(\rho)}{\partial \rho}$ and $-\frac{\partial E_0(\rho)}{\partial \rho}$

First, we compute $\kappa_s'(\rho)$ and set $s=\frac{1}{1+\rho}$ in the result.


```python
d_rho = diff(k, rho)
```

```python
d_rho_rho = d_rho.subs(s, 1/(1+rho))
```

Then, we compute $-E_0'(\rho)$.

```python
e1 = diff(k_rho, rho)
```

By subtracting both and simplifying the expression, we find their relation:

```python
simplify(e1 - d_rho_rho)
```

And, since the result is $0$, this demonstrates that $\kappa_s'(\rho)|_{s=\frac{1}{1+\rho}} = -E_0'(\rho)$.

##$\frac{\partial^2 \kappa_s(\rho)}{\partial \rho^2}$ and  $-\frac{\partial^2 E_0(\rho)}{\partial \rho^2}$

First, we compute $\kappa_s''(\rho)$ and set $s=\frac{1}{1+\rho}$ in the result.

```python
dd_rho = diff(d_rho, rho)
```

```python
dd_rho_rho = dd_rho.subs(s, 1/(1+rho))
```

Then, we compute $-E_0''(\rho)$.

```python
e2 = diff(e1, rho)
```

By subtracting both and simplifying the expression, we find their relation:

```python
simplify(e2 - dd_rho_rho)
```

And, since the result is not $0$ but instead a full expression, this demonstrates that $\kappa_{s}''(\rho)|_{s=\frac{1}{1+\rho}}\neq-E_0''(\rho)$. So, since the relation is not simple, we will use $\kappa_{s}''(\rho)|_{s=\frac{1}{1+\rho}}$.
