```python
from sympy import *
```

In this notebook, $\kappa_{s,r}(\rho)$ and $E_0(\rho, r)$ as well as their derivatives are compared. First, we define the symbols and the symbollic functions:

```python
rho, r, n, x, y, s, alpha, E = symbols('rho r n x y s alpha \mathcal{E}')
Q = Function("Q_{\\text{iid}}")(x)
W = Function("W")(y, x)
```

In the results, $W(y,x)$ corresponds to $W(y|x)$ in the derivations.

##$\kappa_{s,r}^{\text{cc}}(\rho)$

```python
k_cc = log(Integral(summation(Q * exp(r * (pow(x,2) - alpha * E)) * pow(W, 1-s*rho), (x, 0, n)) * pow(summation(Q * exp(r * (pow(x,2) - alpha * E)) * pow(W, s), (x, 0, n)), rho), (y, -oo, oo)))
```

```python
k_cc
```

This result corresponds to $\kappa_{s,r}^{\text{cc}}(\rho)$.

```python
k_cc_rho = k_cc.subs(s, 1/(1+rho))
```

```python
k_cc_rho = simplify(k_cc_rho)
k_cc_rho = simplify(k_cc_rho)
k_cc_rho 
```

This result corresponds to $\kappa_{s,r}^{\text{cc}}(\rho)|_{s=\frac{1}{1+\rho}} = -E_0^{\text{cc}}(\rho,r)$.

##$\frac{\partial \kappa_{s,r}^{\text{cc}}(\rho)}{\partial \rho}$ and $-\frac{\partial E_0(\rho, r)}{\partial \rho}$

First, we compute $\kappa_{s,r}^{\text{cc}}{'}(\rho)$ and set $s=\frac{1}{1+\rho}$ in the result.

```python
d_rho = diff(k_cc, rho)
```

```python
d_rho_rho = d_rho.subs(s, 1/(1+rho))
```

Then, we compute $-E_0^{\text{cc}}{'}(\rho, r)$ w.r.t. $\rho$.

```python
e1 = diff(k_cc_rho, rho)
```

By subtracting both and simplifying the expression, we find their relation:

```python
simplify(e1 - d_rho_rho)
```

And, since the result is $0$, this demonstrates that $\kappa_{s,r}^{\text{cc}}{'}(\rho)|_{s=\frac{1}{1+\rho}} = -E_0^{\text{cc}}{'}(\rho, r)$.

##$\frac{\partial^2 \kappa_{s,r}^{\text{cc}}(\rho)}{\partial \rho^2}$ and  $-\frac{\partial^2 E_0(\rho, r)}{\partial \rho^2}$

First, we compute $\kappa_{s,r}^{\text{cc}}{''}(\rho)$ and set $s=\frac{1}{1+\rho}$ in the result.

```python
dd_rho = diff(d_rho, rho)
```

```python
dd_rho_rho = dd_rho.subs(s, 1/(1+rho))
```

Then, we compute $-E_0^{\text{cc}}{''}(\rho, r)$ w.r.t. $\rho$.

```python
e2 = diff(e1, rho)
```

By subtracting both and simplifying the expression, we find their relation:

```python
simplify(e2 - dd_rho_rho)
```

And, since the result is not $0$ but instead a full expression, this demonstrates $\kappa_{s,r}^{\text{cc}}{''}(\rho)|_{s=\frac{1}{1+\rho}}\neq-E_0^{\text{cc}}{''}(\rho, r)$. So, since the relation is not simple, we will use $\kappa_{s,r}^{\text{cc}}{''}(\rho)|_{s=\frac{1}{1+\rho}}$.
